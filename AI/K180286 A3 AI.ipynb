```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from sklearn import tree

```

```python
#Roll Number: 18k-0286

# if = 6 consider as 5, given in assignment discription 

# as 5 satisfy case 2 
# prime
# odd
# less than 6
# so (5 + 3) / 2 = 4
# a=4, a+2=6, a+3=7
# 4, 6, 7
#Preparing Data
data = load_iris()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.DataFrame(data.target, columns =["Class"])

```

```python
#Tree 1: Option 4
tree1 = DecisionTreeClassifier(min_samples_leaf=4)
tree1.fit(x, y)
predictions1 = tree1.predict(x)
accuracy_tree1 = accuracy_score(y, predictions1)
tree.plot_tree(tree1)
# plt.savefig("tree1.png")
```

```python
#Tree 2: Option 9
tree2 = DecisionTreeClassifier(max_leaf_nodes=6)
tree2.fit(x, y)
predictions2 = tree2.predict(x)
accuracy_tree2 = accuracy_score(y, predictions2)
tree.plot_tree(tree2)
# plt.savefig("tree2.png")
```

```python

#Tree 3: Option 13
tree3 = DecisionTreeClassifier(max_leaf_nodes=7)
tree3.fit(x, y)
predictions3 = tree3.predict(x)
accuracy_tree3 = accuracy_score(y, predictions3)
tree.plot_tree(tree3)
# plt.savefig("tree3.png")
```

```python
#Question 3
print('Accuracy of tree1: ', accuracy_tree1)
print('Accuracy of tree2: ', accuracy_tree2)
print('Accuracy of tree3: ', accuracy_tree3)
```
