```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib as m
import pandas as pd
from scipy.stats import norm
```

```python
data=load_iris()
x,y,features= data['data'],data['target'],data['feature_names']
x=pd.DataFrame(x,columns=features)
x.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train ,y_val= train_test_split(x,y,random_state=50)
```

```python
prob =x_train.groupby(y_train).apply(lambda x : len(x)) /x_train.shape[0]
prob
```

```python
y_pred=[]
means=x_train.groupby(y_train).apply(np.mean)
st_dev=x_train.groupby(y_train).apply(np.std)
for i in range(x_val.shape[0]):
  p={}
  for clas in np.unique(y_train):
    p[clas]=prob.iloc[clas]
    for index , param in enumerate(x_val.iloc[i]):
      p[clas] = p[clas] * norm.pdf(param, means.iloc[clas,index ],st_dev.iloc[clas,index])
      #print (norm.pdf(param, means.iloc[clas,index ],st_dev.iloc[clas,index]))
  y_pred.append(pd.Series(p).values.argmax())#finding class with largest predicted probabilty and appending to y_pred


#y_pred
```

```python
from sklearn.naive_bayes import GaussianNB
model =GaussianNB()
model .fit (x_train,y_train)
predicted= model.predict(x_train)
predicted
```
